```python
import pandas as pd
import matplotlib.pyplot as plt

birds_data = pd.read_csv('~/Downloads/birds.csv')

# Explore the dataset and print first few rows
print(birds_data.head())

#Create a histogram of bird family
plt.figure(figsize=(10, 6))
birds_data['Family'].value_counts().plot(kind='bar', color='blue')
plt.title('Distribution of Bird family')
plt.xlabel('Family')
plt.ylabel('Frequency')
plt.show()

# Example: Create a histogram of bird body mass
plt.figure(figsize=(8, 6))
plt.hist(birds_data['MaxBodyMass'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Bird Sizes')
plt.xlabel('MaxBodyMass')
plt.ylabel('Frequency')
plt.show()
```
